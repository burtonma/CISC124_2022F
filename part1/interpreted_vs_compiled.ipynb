# Interpreted versus compiled languages

One of the main differences between Python and Java is that Python is usually considered to be an interpreted language and Java is considered a compiled language.

## Python as an interpreted language

When programmers are first introduced to Python they usually start by typing Python statements into a program called a Python interpreter that has an interactive prompt. After a statement is typed and the programmer presses the Enter key, the interpreter appears to execute the statement. This form of interaction is called a REPL or read-evaluate-print loop.

As an example, the Python version of the "Hello, World!" program can be typed into an interactive interpreter and it runs as soon as the user presses the Enter key.

````java
%%python
print("Hello, world!")
````

Although the interpreter appears to read the text from the prompt and execute it, what it actually does is compile the text into Python bytecode; however, it does so silently without any input from the programmer. The bytecode is then run on a Python virtual machine. This process should sound strikingly similar to that described in the [The Java Virtual Machine and Java compiler](./java_virtual_machine_and_java_compiler.ipynb#notebook_id).

## Java as a compiled language

Java is considered to be a compiled language because the compilation process is usually an explicit step; however, the bytecode generated by the compiler is interpreted by the JVM.

Let's revisit the "Hello, world!" program shown below and experience how such a program would be created, compiled, and run in the early days of the Java language:

```java
class HelloWorld {

    public static void main(String... args) {
        System.out.println("Hello, world!);
    }
}
```

### Manually compiling and running the "Hello, world!" program

If you have a working Java Development Kit (JDK) environment you can create the Java file `HelloWorld.java`, compile it to bytecode using the Java compiler, and then run the resulting bytecode on a JVM using the following steps. A video of the process is shown after the instructions.

1. Create a plain text file named `HelloWorld.java`
2. Edit the file so that it contains the Java code shown above
3. Save the file
4. Open a command-line interpreter or terminal emulator for your operating system in the directory containing your `HelloWorld.java` file
5. Compile `HelloWorld.java` to bytecode using the Java compiler
    + The standard compiler is called `javac` in Linux and MacOS and `javac.exe` in Windows.
    + In Linux or MacOS you would type the following at the command prompt and then press the Enter key:
    ```
    javac HelloWorld.java
    ```
    + In Windows you would type the following at the command prompt and then press the Enter key:
    ```
    javac.exe HelloWorld.java
    ```
6. Run the bytecode using the Java application launcher
    + The standard Java application launcher is called `java` in Linux and MacOS and `java.exe` in Windows.
    + In Linux or MacOS you would type the following at the command prompt and then press the Enter key:
    ```
    java HelloWorld
    ```
    + In Windows you would type the following at the command prompt and then press the Enter key:
    ```
    java.exe HelloWorld
    ```
